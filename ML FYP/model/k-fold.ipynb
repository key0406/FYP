```python
import numpy as np
import pandas as pd
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
import optuna
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, mean_squared_error
import statistics
import joblib

# 
df = pd.read_csv(R"C:\Users\User\Desktop\FYPML\model\n=3500.csv", encoding="932", header=0)
# 
feature_names = [ "P1", "P2","P3", "P4","f1", "f2", "f3", "f4","q1", "q2", "q3", "q4", "Pain", "Function", "QOL"]
#feature_names = ['sex','month','Age', 'period','b_Pain','b_Function','b_QOL','prp_count']


# 
if 'sex' in feature_names:
    df['sex'] = df['sex'].astype('category').cat.codes

# 
X = df[feature_names].values
y = df['score'].values

# 
num_folds = 5
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# 
pred_list = []
r2_scorelist = []
y_test_list = []
evals_result_list = []
rmse_list = []  # foldvalid RMSE

fold = 1
for train_index, test_index in kf.split(X, y):
    X_train_fold, X_test_fold = X[train_index], X[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]

    # Optunaobjective
    def objective(trial):
        params = {
            "objective": "reg:squarederror",
            "eval_metric": "rmse",
            "booster": 'gbtree',
            "max_depth": trial.suggest_int("max_depth", 1, 5),
            "max_leaves": trial.suggest_int("max_leaves", 1, 30),
            "eta": trial.suggest_loguniform("eta", 1e-4, 1),
            "seed": trial.suggest_int("seed", 100, 5000),
            "n_estimators": trial.suggest_int("n_estimators", 1, 10),
            "alpha": trial.suggest_int("alpha", 0, 100),
            "lambda": trial.suggest_int("lambda", 0, 100)
        }

        model = xgb.XGBRegressor(**params)
        model.fit(X_train_fold, y_train_fold)
        pred = model.predict(X_test_fold)
        score = np.sqrt(mean_squared_error(y_test_fold, pred))  # RMSE 
        return score

    # Optuna
    study = optuna.create_study(direction="minimize")
    study.optimize(objective, n_trials=100, timeout=600)

    # DMatrix
    dtrain = xgb.DMatrix(X_train_fold, y_train_fold, feature_names=feature_names)
    dvalid = xgb.DMatrix(X_test_fold, y_test_fold, feature_names=feature_names)

    # 
    params = {
        "objective": "reg:squarederror",
        "eval_metric": "rmse",
        'booster': 'gbtree',
        'eta': study.best_params['eta'],
        'max_depth': study.best_params['max_depth'],
        'max_leaves': study.best_params['max_leaves'],
        'seed': study.best_params['seed'],
        'n_estimators': study.best_params['n_estimators'],
        'alpha': study.best_params['alpha'],
        'lambda': study.best_params['lambda']
    }

    evals_result = {}
    model = xgb.train(
        params=params,
        dtrain=dtrain,
        evals=[(dtrain, "train"), (dvalid, "valid")],
        num_boost_round=2000,
        early_stopping_rounds=10,
        evals_result=evals_result,
        verbose_eval=False
    )

    # RMSE
    train_rmse = evals_result['train']['rmse'][0]
    valid_rmse_list = evals_result['valid']['rmse']
    min_valid_rmse = min(valid_rmse_list)

    print(f"Fold {fold}")
    print(f"Initial train RMSE: {train_rmse}")
    print(f"Minimum valid RMSE: {min_valid_rmse}\n")

    # 
    pred = model.predict(xgb.DMatrix(X_test_fold, feature_names=feature_names))
    pred_list.append(pred)
    r2_scorelist.append(r2_score(y_test_fold, pred))
    y_test_list.append(y_test_fold)
    evals_result_list.append(evals_result)
    
    # RMSE
    rmse_list.append(min_valid_rmse)

    fold += 1

# 
for i in range(5):
    sns.set(font=['Meiryo', 'Yu Gothic'])
    plt.scatter(y_test_list[i], pred_list[i], alpha=0.5)
    plt.plot(np.linspace(min(y_test_list[i]), max(y_test_list[i])),
             np.linspace(min(y_test_list[i]), max(y_test_list[i])), "red")
    plt.xlabel('KOOS')
    plt.ylabel('KOOS')
    plt.title(f'Fold {i+1}')
    plt.show()
    print(f"R (Fold {i+1}): {r2_scorelist[i]}")

# 
_, ax = plt.subplots(figsize=(12, 8))
xgb.plot_importance(model,
                    ax=ax,
                    importance_type='gain',
                    show_values=True)
plt.show()

# loss
for i in range(5):
    plt.plot(evals_result_list[i]['train']['rmse'], label='train rmse')
    plt.plot(evals_result_list[i]['valid']['rmse'], label='valid rmse')
    plt.grid()
    plt.legend()
    plt.xlabel('rounds')
    plt.ylabel('rmse')
    plt.title(f'Loss Curve for Fold {i+1}')
    plt.show()

# rmse
for i in range(5):
    print(f"Fold {i+1}")
    print("train rmse:", min(evals_result_list[i]['train']['rmse']))
    print("valid rmse:", min(evals_result_list[i]['valid']['rmse']))
    print()

# R
all_preds = np.concatenate(pred_list)
all_y_test = np.concatenate(y_test_list)
overall_r2 = r2_score(all_y_test, all_preds)

# RMSE
mean_min_rmse = np.mean(rmse_list)

# 
print("R scores for each fold:", r2_scorelist)
print("Mean R score across folds:", np.mean(r2_scorelist))
print("Overall R score:", overall_r2)

print("Minimum valid RMSE for each fold:", rmse_list)
print("Mean of minimum valid RMSE across folds:", mean_min_rmse)

# 
plt.scatter(all_y_test, all_preds, alpha=0.5)
plt.plot([min(all_y_test), max(all_y_test)],
         [min(all_y_test), max(all_y_test)], color="red")
plt.xlabel("Actual KOOS Score")
plt.ylabel("Predicted KOOS Score")
plt.title("Scatter Plot for Overall Prediction")
plt.show()
joblib.dump(model, "KOOS_prediction_3500_model.pkl")
print("Model saved")
```
